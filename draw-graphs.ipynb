```python
def draw_charts(files):
    ncols=3
    fig, axes = plt.subplots(nrows=round(len(files)/ncols), ncols=ncols, figsize=(20, 15))
    plt.subplots_adjust(hspace=1., wspace=1.)
    fig.suptitle('Hierarchies per time step')
    
    for f, ax in zip(files, axes.flatten()):
        ax.set_title('time step ' + f.split('.')[0])
        g=nx.read_adjlist(os.path.join('sim-graphs', f))
        nx.draw_networkx(g, with_labels=True, ax=ax)
    plt.tight_layout()
    plt.plot()
```

```python
gfiles = os.listdir('sim-graphs')
gfiles = sorted(gfiles, key=lambda x: int(x.split('.')[0]))
draw_charts(gfiles[1:])
```
